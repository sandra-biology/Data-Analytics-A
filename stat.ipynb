```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

a = 3
b = 5
print(f'a: {a}, b: {b}')

x = np.arange(10)
y = a * x + b

sigma = 3

np.random.seed(1)

dy = np.random.normal(0, sigma, len(x))
y = y + dy
print(f'y: {y}')

ffit = lambda x, a, b: a*x + b

popt, pcov = curve_fit(ffit, x, y)
perr = np.sqrt(np.diag(pcov))

print(f'popt: {popt}')
print(f'pcov:\n{pcov}')
print(f'perr: {perr}')

yfit = ffit(x, popt[0], popt[1])

plt.figure()
plt.title(f'a_fit: {popt[0]:0.2f}, b_fit: {popt[1]:0.2f}')
plt.plot(x, y, 'o')
plt.plot(x, yfit)
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

a = 3
b = 5
print(f'a: {a}, b: {b}')

x = np.arange(10)
y = a * x + b

sigma = 3

np.random.seed(1)

dy = np.random.normal(0, sigma, len(x))
y = y + dy
print(f'y: {y}')

ffit = lambda x, a, b: a*x + b

popt, pcov = curve_fit(ffit, x, y)
perr = np.sqrt(np.diag(pcov))

print(f'popt: {popt}')
print(f'pcov:\n{pcov}')
print(f'perr: {perr}')

a_fit, b_fit = popt

yfit = ffit(x, *popt)

chi2 = np.sum( ((a_fit * x + b_fit - y) / sigma)**2 )
ndf = len(x) - 2  # Number Degrees of Freedom

plt.figure()
plt.title(f'a_fit: {a_fit:0.2f}, b_fit: {b_fit:0.2f}, $\chi^2/ndf$: {chi2 / ndf:0.2f}')
plt.plot(x, y, 'o')
plt.plot(x, yfit)
plt.grid()
```

# Percentile

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

data = np.random.normal(size=10000)
# print(y)
```

```python
plt.figure()
y, x, _ = plt.hist(data, bins=40)
plt.grid()

print(f'len(x): {len(x)}, len(y): {len(y)}')  # len(y) = 40, len(x) = 41

x = (x[1:] + x[:-1]) / 2  # find centers of the bins
print(f'len(x): {len(x)}, len(y): {len(y)}')

plt.figure()
plt.plot(x, y, '.')
plt.grid()
```

```python
y_sum = 0
y_cum = np.zeros(len(y))
for i in range(len(y)):
    y_sum += y[i]
    y_cum[i] = y_sum

plt.figure()
plt.title('Cumulative distribution')
plt.plot(x, y_cum, '.')
plt.grid()
```
