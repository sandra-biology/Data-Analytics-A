```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

student_performance_df = pd.read_csv(r'C:\\Users\\CICD STUDENT\\Desktop\\Data-Analytics-A\\Datasets\\exams.csv')
student_performance_df.head()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='parental level of education', y='math score', data=student_performance_df)
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='parental level of education', y='math score', data=student_performance_df, hue='gender')

# move legend outside of plot area
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# changing orientation

# NOTICE HOW WE HAVE TO SWITCH X AND Y FOR THE ORIENTATION TO MAKE SENSE!
sns.boxplot(x='math score', y='parental level of education', data=student_performance_df, orient='h')
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x='parental level of education', y='math score', data=student_performance_df)
```

```python
# changing width

plt.figure(figsize=(10, 4))
sns.boxplot(x='parental level of education', y='math score', data=student_performance_df, width=0.3)
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df)
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, hue='gender')
```

#### Violinplot parameters

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, hue='gender', split=True)
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, inner=None)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, inner='box')
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, inner='quartile')
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(x='parental level of education', y='math score', data=student_performance_df, inner='stick')
```

```python
# simply switch the continuous variable to y and the categorical to x

sns.violinplot(x='math score', y='parental level of education', data=student_performance_df)
```

```python
# bandwidth is similar to bandwidth argument in kde plot
sns.violinplot(x='math score', y='parental level of education', data=student_performance_df, bw=0.1)
```

#### Jointplot

```python
sns.jointplot(x='math score', y='reading score', data=student_performance_df)
```

```python
plt.figure(figsize=(6, 4))
sns.jointplot(x='math score', y='reading score', data=student_performance_df, kind='hex')
```

```python
sns.jointplot(x='math score', y='reading score', data=student_performance_df, kind='kde')
```
