```python
import numpy as np

x = np.arange(0,10)
y = 2 * x
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.xlabel('x axis title here: x')
plt.ylabel('y axis title here: y')
plt.title('y = 2x')
plt.show()
```

```python
plt.plot(x, y) 
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.xlim(0,6) # Lower Limit, Upper Limit
plt.ylim(0,12) # Lower Limit, Upper Limit
plt.show() # Required for non-jupyter users , but also removes Out[] info
```

Object-oriented

```python
x = np.arange(10)
y = x

# create figure object
fig = plt.figure()

# add axes to that figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# plot on that axes
axes.plot(x, y)
```

```python
a = np.linspace(0, 10, 11)
b = a ** 2
print(a)
print(b)
```

```python
# create Figure (empty canvas)
fig = plt.figure()

# add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# plot on that set of axes
axes.plot(a, b)
axes.plot(x, y)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

Adding multiple axes to figure

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # Smaller figure

# larger axes figure 1
axes1.plot(a, b)

# use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# insert axes figure 2
axes2.plot(x,y)
axes2.set_title('Small Figure')
```

```python
# moving small figure and editing its parameters

fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure

# larger axes figure 1
axes1.plot(a, b)

# use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# insert axes figure 2
axes2.plot(x,y)
axes2.set_xlim(0,10)
axes2.set_ylim(0,20)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In')
```

```python
# putting small figure outside the big figure

# create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Full figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure
axes3 = fig.add_axes([1,1,0.25, 0.25]) # Starts at top right corner!

# larger Figure Axes 1
axes1.plot(a, b)

# use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

# insert Figure Axes 3
axes3.plot(a,b)
```

## Subplots

```python
x = np.arange(0,10)
y = 2 * x

# use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# now use the axes object to add stuff to plot
axes.plot(x, y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); #; hides Out[]
```

```python
# empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes is an array of axes to plot on
axes
axes.shape
```

```python
# empty canvas of 2 by 2 subplots
fig, axes = plt.subplots(nrows=2, ncols=2)

print(axes)
axes.shape
```

Plotting on axes objects

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for axis in axes:
    axis.plot(x, y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

Use fig.tight_layout() or plt.tight_layout() to not have overlapping x axis labels

```python
# parameters on subplots

fig,axes = plt.subplots(nrows=3,ncols=3)

# set axes parameters first

# parameters at the axes level
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')


axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

axes[2][2].plot(y,x)
axes[1][0].plot(b,a)

# then set overall figure parameters

# parameters at the Figure level
fig.suptitle("Figure Level",fontsize=16)  # add a centered title to the figure

plt.tight_layout()

plt.show()
```
