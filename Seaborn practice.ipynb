### Scatterplot

```python
import seaborn as sns

sns.scatterplot(x='salary', y='sales', data=office_sales_df)
```

We can connect Seaborn plots to MatPlotLib's figsize

```python
import matplotlib.pyplot as plt

# plt.figure(figsize=(12, 8), dpi=100)
plt.figure(figsize=(8, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=office_sales_df)
```

```python
# adding hue parameter

plt.figure(figsize=(8, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=office_sales_df, hue='division')
```

```python
# adding palette parameter

plt.figure(figsize=(10, 6), dpi=100)
sns.scatterplot(x='salary',y='sales',data=office_sales_df,hue='work experience',palette='viridis')
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.scatterplot(x='salary',y='sales',data=office_sales_df,hue='work experience',palette='Accent')
```

Scatterplot parameters
- size allows us to size based on another column
- **use *s=* if you want to change the marker size to be some uniform integer value**
- style
    - Automatically choose styles based on another categorical feature in the dataset. Optionally use the **markers=** parameter to pass a list of marker choices based off matplotlib, for example: ['*','+','o']

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=office_sales_df, size='work experience')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=office_sales_df, s=200)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='salary',y='sales',data=office_sales_df,s=200,linewidth=0,alpha=0.5)
```

```python
plt.figure(figsize=(12,8))
# sometimes its nice to do BOTH hue and style off the same column
sns.scatterplot(x='salary',y='sales',data=office_sales_df,style='level of education',hue='level of education',s=100)
```

## Kernel density estimation plots

The KDE plot maps an estimate of a probability *density* function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

```python
import numpy as np

np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
sample_ages_df = pd.DataFrame(sample_ages, columns=['age'])
sample_ages_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data=sample_ages_df, x='age', bins=10, kde=True)
```

```python
sns.kdeplot(data=sample_ages_df, x='age')
```

```python
# cutting off the KDE
sns.kdeplot(data=sample_ages_df, x='age', clip=[0, 100])
```
