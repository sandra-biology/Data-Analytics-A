```python
# %matplotlib qt
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

mean = 5
sigma = 1  # standard deviation
a = np.random.normal(mean, sigma, 100)

plt.figure()
plt.title(f'$\mu$ = {mean}, $\sigma$ = {sigma}')
plt.hist(a)
plt.grid()
```

```python
np.random.seed(1)

mean = 5
sigma = 1  # standard deviation
a = np.random.normal(mean, sigma, 100)

nbins = 20

plt.figure()
plt.title(f'$\mu$ = {mean}, $\sigma$ = {sigma}')
plt.hist(a, bins=nbins)
plt.grid()
```

```python
np.random.seed(1)

mean = 5
sigma = 1  # standard deviation
a = np.random.normal(mean, sigma, 100)

nbins = 20
bins = np.linspace(0, 10, nbins + 1)
print(f'bins: {bins}')

plt.figure()
plt.title(f'$\mu$ = {mean}, $\sigma$ = {sigma}')
plt.hist(a, bins=bins)
plt.grid()
```
