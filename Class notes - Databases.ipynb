### Introduction to databases
- components
    - data
    - tables: data is organized into tables
        - each row is separate record
    - schema: defines structure of the database, specifying tables, their fields, relationship between them
    - querries: requests to retrieve, update, or manipulate data based on specific criteria
    - ACID properties: atomicity, consistency, isolation, durability
- advantages
    - data integrity and security
    - efficient data retrieval
    - data consistency
    - scalability
    - data backup and recovery
    - data analysis
- relational databases
    - use structured methodology of storing and organizing data
    - SQL (language) is used to manage and manipulate these databases
    - primary vs foreign keys
        - primary key is a unique identifier for each record (row) in a table
            - unique, not null (valid), unchangeable
            - defined when creating a table
        - foreign key: column in one table that refers to the primary key in another table
            - establishes a relationship between two tables, creating a link between the data in both tables
            - references primary key, can have null values
- SQL
    - standard language for interacting with relational databases
    - allows you to define structure and schema of a database using DDL commands
        - can create, alter,, and drop database objects
    - provides commands for manipulating data stored in the database
        - operations such as inserting, updating, deleting, querying data using DML commands
    - powerful querying capabilities allow you to filer, sort, group, and join data from different tables to get meaningful insights and reports
    - enables you to define various constraints on the data to maintain data integrity
    - supports transaction management
    - offers mechanism to grant and revoke priviledges on database objects
- NoSQL databases
    - provide a non-tabular and flexible approach for storign and managing data
    - designed to handle vast amounts of unstructured or semi-structured data
    - use variety of data models
    - typically have a dynamic or schema-less nature
    - designed to be horizontally scalable
    - often follow the BASE (Basically Available, Soft state, Eventually consistent) model
    - types of NoSQL databases
        - document databases
        - key-value databases
        - wide-column (column-family) databases
        - graph databases

### Introduction to big data
- big data characterized by 4 Vs: volume, velocity, variety, and sometimes veracity
- can reveal patterns and associations, especially relating to human behavior and interactions
- Hadoop
    - open-source framework designed to store, process, analyze large amounts of data in a distributed computing environment
    - key components
        - HDFS (Hadoop Distributed File System)
        - MapReduce: programming model and processing engine for distributed data processing
        - YARN (Yet Another Resource Negotiator): resource management layer that schedules and allocates resources to different applications running on the cluster
            - allows for multi-tenancy
            - enables data processing frameworks to run alongside MapReduce
    - MapReduce
        - Map: data divided into chunks, each chunk is processing independently by different mappers
        - Reduce: each reducer processes a subset of sorted intermediate data, performing data aggregation, summarization, or other computations, producing the final output
        - key features: fault tolerance, scalability, data locality
- Spark
    - open-source, distributed computing system designed for fast and scalable data processing
    - differences with Hadoop
        - Hadoop well suited to batch processing and large scale data processing, Spark excels at real-time processing, iterative algorithms, interactive data analysis
        - Spark provides built-in support for stream processing, making it more convenient for handling real-time streaming data
    - real-world examples
        - genomics research
            - facilitates analysis of large-scale genomics data
            - helping researchers gain insights into genetic variantions and associations with disease
        - drug discovery
            - process and analyze molecular data
            - supporting drug discovery and developing efforts
    - can run on Hadoop or standalone clusters

### SQL commands
- select: specifies which columns to retrieve
    - from clause: specifies table to retrieve data from
    - where: optional, allows us to specify filtering conditions
    - order by clause: specifies order in which to show data
    - grouby clause:
    - can create new column in new table in the same query just by using summation
- use master; to query entire database
- insert: adds new records to a table in a relational database
    - essential for adding new data to table
- update statement modifies existing record
    - set clause specifies columns to be modified and their new values
    - where clause (optional) defines condition that determines which rows will be updated
        - highly recommended
        - can specify multiple conditions in the same where clause, just need to separate conditions using and
            - can use or condition like in Python
- delete: removes records from table
    - table name
    - where specifies conditions under which records will be removed
        - without it all rows get deleted, resulting in large data loss
- join: combines data from multiple tables based on related columns or conditions
    - types: inner join, left join, right join, full join
    - inner join: returns only the rows where there is a match in both tables
    - left join: keeps left table (table before the join statement) intact, puts in null where right table doesn't have a match
    - right join: keeps rihgt table intact, puts in null where left table doesn't have a match
    - full join: returns all rows where there is a match in either left or right table
